#### 코드 테이블 만들기

```python
# 엑셀에서 작업해도 되나 python을 이용해서 작업한다면 복잡하나 단순작업이 줄어듦
# 특정 excel의 여러 sheet를 읽어 올 때 시간 절약
xls = pd.ExcelFile('../data/MCR2019_최종.xlsx')
code1 = pd.read_excel(xls, '문항정보')
code2 = pd.read_excel(xls, '보기정보')
code1.head(3)
code2.head(3)
```

```python
# 빈 값의 칸을 앞 줄 값으로 대체 





code2.head(3)
```

```python
code1.to_pickle('../code/MCR2019_code1.pkl')
code2.to_pickle('../code/MCR2019_code2.pkl')
```

```python
# code1과 code2 merge후 regex를 이용해서 대상 컬럼 선택하고 컬럼명 짧게 변경
code3 = code1.merge(code2, on='항목ID')[['항목명','항목코드','항목코드명']]
code3 = code3[code3['항목명']
              .str
              .contains('성별|1세|지역|직업|가구 월평균|최종학력'
                        +'|주거 형태|결혼 여부|자녀 총 수'
                        +'|인터넷 이용 ?시간|즐겨 보는 TV|드라마 장르-내용'
                        +'|제품/브랜드 구매의향')].reset_index(drop=True)
```

```python
code3['컬럼명'] = (code3['항목명']
                  .str.replace('가구 월평균 | 이용시간| 인터넷 이용 시간| 하루 평균 ','')
                  .str.replace('즐겨 보는 | 프로그램 장르|TV 시청 시 즐겨 보는 국내 ','')
                  .str.replace(' 장르-내용|PPL\(간접광고\) 평가 - 제품/브랜드','')
                  .str.replace('TV$','TV_0')
                  .str.replace('드라마$','드라마_0')
                  .str.replace('\.','_')
                  .str.replace('\(.*\)| ',''))

code3.to_pickle('../code/code3.pkl')
```

#### json형태의 코드 테이블 이용하기

```python
# json file을 dict로 만들어 이용하기
import json
choices_org = json.load(open('../code/choices.json', 'r'))
```

```python
choices_org['지역']
```

```python
choices = {}
for choice in choices_org:
    temp = {}
    for item in choices_org[choice]:
        temp[int(item)] = choices_org[choice][item]
    choices[choice] = temp

choices['지역']
```

```python
col = '지역'
df[col].value_counts().sort_index().rename(col).rename(index=choices[col])
```
