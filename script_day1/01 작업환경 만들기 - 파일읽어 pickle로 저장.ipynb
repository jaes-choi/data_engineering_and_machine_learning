```python
# 컬럼명 변경 - 컬럼명이 너무 길어서 축약하기로 함
# 명시적으로 컬럼명을 적고 df.columns에 대입
cols_new = ['성별','연령','지역','직업','소득','최종학력','주거형태','결혼여부','자녀총수',
            '평일인터넷','토요일인터넷','일요일인터넷','평일모바일','토요일모바일','일요일모바일',
            'TV_0','TV_1','TV_2','TV_3','TV_4','TV_5','TV_6','TV_7','TV_8',
            'TV_9','TV_10','TV_11','TV_12','TV_13','TV_14','TV_15','TV_16',
            '드라마_0','드라마_1','드라마_2','드라마_3','드라마_4','드라마_5',
            '드라마_6','드라마_7','드라마_8','드라마_9','드라마_10',
            '드라마_11','드라마_12','드라마_13','드라마_14','드라마_15',
            '구매의향'] 
df.columns = cols_new
df.to_pickle('../working/MCR2019_selected.pkl')
```

```python
# 4가지 기초정보를 dataframe으로 만들기 - 같은 성격의 컬럼인데 다른 dtype 발견
pd.DataFrame({'dtypes':df.dtypes,
              'nunique':df.nunique(),
              'notnull':df.notnull().sum(),
              'isnull':df.isnull().sum()})
```

```python
# 'TV'와 '드라마'가 같은 성질의 컬럼임에도 다른 data type임을 확인
# 'TV'와 '드라마'가 컬럼명에 있는 컬럼들의 unique한 값 인쇄
for col in df.columns:
    if ('TV' in col) | ('드라마' in col):
        print(col, df[col].unique())
```

```python
# '#NULL!'값을 null로 변경
import numpy as np
df = df.reset_index(drop=True)
for col in df.columns:
    if ('TV' in col) | ('드라마' in col):
        df[col] = df[col].apply(lambda x: np.nan if x=='#NULL!' else x)
```

```python
# 명백한 오류가 제거된 dataframe 저장
df.to_pickle('../working/MCR2019_null_removed.pkl')
```

```python
# 성별 ~ 자녀총수 컬럼들에 대한 확인 --> dtype이 int64: value_counts로 확인
for col in df.columns:
    print('='*30)
    print(df[col].value_counts().sort_index())
    if col=='자녀총수':
        break 

# 모든 결과가 숫자로 인코딩되어 있음 발견 --> 어떻게 해야 할까?
```

```python
# 정규표현식 사용해서 컬럼명 변경 방법
cols_new = (df.columns
              .str.replace('가구 월평균 | 이용시간| 인터넷 이용 시간| 하루 평균 ','')
              .str.replace('즐겨 보는 | 프로그램 장르|TV 시청 시 즐겨 보는 국내 ','')
              .str.replace(' 장르-내용|PPL\(간접광고\) 평가 - 제품/브랜드','')
              .str.replace('TV$','TV_0')
              .str.replace('드라마$','드라마_0')
              .str.replace('\.','_')
              .str.replace('\(.*\)| ','')
            )
cols_new
```

```python
# null값 시각화
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

s_temp = df.notnull().sum()
s_temp.plot.bar(rot='45', figsize=(20,4))
max_val = s_temp.max()
for idx, item in enumerate(s_temp):
    plt.text(idx-0.4, max_val*1.1, item, rotation=45)
```
