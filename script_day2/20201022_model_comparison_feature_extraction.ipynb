# <font color='red'> 다양한 특성 변수 추출 방법
https://scikit-learn.org/stable/modules/feature_selection.html

- 필터 방법(Filter method)은 독립변수의 분산, 종속변수와의 상관관계, 카이제곱값(범주형 변수의 경우) 등 단일 변수 통계량을 기준으로 한다. 
- 래퍼 방법(Wrapper method)은 일부 변수를 사용하여 모델에서 도출된 추론으로 특성변수를 추가 제거하는 반복적 방법이다. 
- 임베디드 방법(Embedded method)은 필터와 래퍼 방법을 결합한 로지스틱 회귀, 의사결정트리 등을 적용하여 특성변수를 추출한다.

```python
df = pd.read_excel('../data/encoded_data.xlsx')
```

```python
X = df.drop(['구매'], axis=1)
y = df['구매']
X.shape, y.shape
X.head()
```

## 특성변수 개수 설정

```python
feature_name = X.columns
best_features = 25
```

## 분산 추출 방법(Low variance)

```python
from sklearn.feature_selection import VarianceThreshold
matrix = np.array([[0, 0, 1], [0, 1, 0], [1, 0, 0], [0, 1, 1], [0, 1, 0], [0, 1, 1]])
data = pd.DataFrame(matrix)
data.loc['variance'] = data.var()
data.loc['number of containing zero'] = ((data[0:6] == 0)*1).sum()
data.loc['probability of containing zero'] = data.loc['number of containing zero']/matrix.shape[0]
data.round(2)

p = 0.8   # 변수가 '0'을 가질 확률
threshhold = p * (1 - p)
print(f'threshhold is {threshhold}.')
sel = VarianceThreshold(threshold=(p * (1 - p)))
sel.fit_transform(matrix)
```

```python
def var_selector(X, p):
    threshhold = p * (1 - p)
    sel = VarianceThreshold(threshold=(p * (1 - p)))
    var_selector = VarianceThreshold(threshold=(p * (1 - p)))
    var_selector.fit(X)    
    var_support = var_selector.get_support()
    var_feature = X.iloc[:, var_support].columns.tolist()

    return var_feature, var_support

var_feature, var_support = var_selector(X, 0.8) # threshhold
','.join(var_feature)
''.join(str(var_support))
```

## 상관계수 (Pearson correlation)

```python
def corr_selector(X, y):

    cor_list = []
    for i in X.columns:
        cor = np.corrcoef(X[i], y)[0, 1]
        cor_list.append(cor)
    cor_feat = X.iloc[:, np.argsort(np.abs(cor_list))[best_features*-1:]].columns.tolist()
    cor_support = [True if i in cor_feat else False for i in feature_name]

    return cor_feat, cor_support 
```

```python
corr_feature, corr_support = corr_selector(X, y)
'/'.join(corr_feature)
''.join(str(corr_support))
```

## 카이제곱 (Chi-square)

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.preprocessing import MinMaxScaler

def Chi_selector(X, y):
    chi_selector = SelectKBest(chi2, k=best_features)
    X_norm = MinMaxScaler().fit_transform(X)
    chi_selector.fit(X_norm, y)
    chi_support = chi_selector.get_support()
    chi_feature = X.loc[:, chi_support].columns.tolist()

    return chi_feature, chi_support

chi_feature, chi_support = Chi_selector(X, y)
','.join(chi_feature)
''.join(str(chi_support))
```

## RFE (Recursive Feature Elimination)
기계학습의 로지스틱 회귀를 모든 변수를 넣어 적용/ 학습을 진행하여 중요도가 낮은 변수를 하나씩 제거하는 모듈이다

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
def Wrapper_RFE(X, y):
    rfe_selector = RFE(estimator=LogisticRegression(),
                        n_features_to_select=best_features, step=20)
    X_norm = MinMaxScaler().fit_transform(X)
    rfe_selector.fit(X_norm, y)
    rfe_support = rfe_selector.get_support()
    rfe_feature = X.loc[:,rfe_support].columns.tolist()
    
    return rfe_support, rfe_feature

rfe_support, rfe_feature = Wrapper_RFE(X, y)
','.join(rfe_feature)
''.join(str(rfe_support))
```

## 로지스틱 모델

```python
from sklearn.feature_selection import SelectFromModel

def logistic_regression(X, y):
    embeded_lr2_selector = SelectFromModel(LogisticRegression(), # default l2
                                           threshold='median') 
    X_norm = MinMaxScaler().fit_transform(X)
    embeded_lr2_selector.fit(X_norm, y)
    embeded_lr2_support = embeded_lr2_selector.get_support()
    embeded_lr2_feature = X.loc[:,embeded_lr2_support].columns.tolist()
    
    return embeded_lr2_support, embeded_lr2_feature

embeded_lr2_support, embeded_lr2_feature = logistic_regression(X, y)
','.join(embeded_lr2_feature)
len(embeded_lr2_feature)
```

## 랜덤포레스트 모델

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier

def random_forest(X, y):
    embeded_rf_selector = SelectFromModel(RandomForestClassifier(), 
                                          threshold='median')
    embeded_rf_selector.fit(X, y)    
    embeded_rf_support = embeded_rf_selector.get_support()
    embeded_rf_feature = X.loc[:,embeded_rf_support].columns.tolist()
    return embeded_rf_support, embeded_rf_feature

embeded_rf_support, embeded_rf_feature = random_forest(X, y)
'/'.join(embeded_rf_feature)
```
